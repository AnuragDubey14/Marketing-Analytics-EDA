# The Marketing Manager wants to know the 'Age' of the customers. Extract the age feature from the given dataset and display the statistical summary of the age? (Statistical Summary: Show the dispersion of the data in the form of graphs. Perform outlier treatment Categorize data in Age Groups)

```python
import datetime
current_year= datetime.datetime.now().year
df2['Age']=current_year - df2['Year_Birth']
df2
```

```python
#Creating Age Group based on Various Ages
def agegrp(age):
  if age<10:
    return 'Kid'
  elif age>10 and age<=18:
    return 'Teenager'
  elif age>18 and age<45:
    return 'Adult'
  elif age>=45 and age<65:
    return 'Middle Age'
  elif age>=65:
    return 'Senior Citizen'

df2['Agegroup']=df2['Age'].apply(agegrp)
df2
```

```python
x=sns.countplot(x=df2['Agegroup'])
for label in x.containers:
  x.bar_label(label)
plt.title("People According to various age group")
plt.show()
```

From above diagram we see that there is 1237 customer from Middle age and 515 customer from Senior Citizen and 476 customer from Adult catgory.

# Perform Correlation Analysis and write your key inferences.

```python
x=df2.corr().abs().unstack().sort_values(ascending=False)
x = x[x != 1].drop_duplicates()
top_5_corr=pd.DataFrame(x.head(5))
top_5_corr.reset_index(inplace=True)
```

```python
top_5_corr
```

```python
correlation_matrix = top_5_corr.pivot('level_0', 'level_1', 0)


plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix Heatmap')
plt.show()
```
